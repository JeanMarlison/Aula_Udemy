```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
# Aqui fizemos a leitra do nosso DataSet e add em gas
```

```python
plt.plot(gas['Year'], gas['Australia']) # Aqui foi com só Anos e um País
```

Fazendo com mais Paises

```python
plt.plot(gas['Year'], gas['Australia'])
plt.plot(gas['Year'], gas['Italy'])
plt.plot(gas['Year'], gas['USA'])


# Opcional 
plt.show()
```

Nos graficos acima podemos ver quem paga mais cara no Gas de Combustivel

### Adicionar Legenda e Ticks

```python
plt.title('Valor do Combustivel em US$')

plt.plot(gas['Year'], gas['Australia'], label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], label = 'Italy')
plt.plot(gas['Year'], gas['USA'], label = 'USA')
plt.plot(gas['Year'], gas['Japan'], label = 'Japan')

plt.xticks(gas['Year'][::2]) # Estou alterando o Eixo x [ : : ] [Incio : Fim : De Tantos em Tantos]
# Neste caso ficou do [::2] Do incio ao fim de dois em dois anos

plt.legend()
plt.show()
```

### Formatar Label e Cores

```python
plt.title('Valor do Combustivel em US$')

plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], 'y*--',  label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b+--',label = 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g>--', label = 'Japan')

# Na Australia  colocamos também cor, o ponto (.) e o tipo da linga que é um taço (-) mas poderia ser mais (---). 
# A estrutura ficou  ,'r.-',
# Vamos ver na Italy , 'y*--',
# USA , 'b+--',
# Japan , 'g>--',

plt.xticks(gas['Year'][::2]) 

plt.xlabel('Years')  # Legenda de X que vai ser Year(Anos)
plt.ylabel('US$') # Legenda de Y que vai ser US$ (valor em dolar)

plt.legend()
plt.show()
```

### Adicionando com For Loop

```python
plt.title('Valor do Combustivel em US$')

lista_Para_Contar = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'germany']

for country in gas:
    if country in lista_Para_Contar:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xticks(gas['Year'][::2]) 

plt.xlabel('Years')
plt.ylabel('US$') 
plt.legend()
plt.show()
```

### Salvando Graficos do MatPlotLib

Como estou usando o pyplot do MatPlotLib, é ir na documentação e procurar pelo pyplot 

```python
plt.title('Valor do Combustivel em US$')

lista_Para_Contar = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'germany']

for country in gas:
    if country in lista_Para_Contar:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xticks(gas['Year'][::2]) 

plt.xlabel('Years')
plt.ylabel('US$') 
plt.legend()

# Acredite Isto está na Documentação
# Tem que ser antes do show() ele limpa e se colocar depois, você vai salver um arquivo em Branco
plt.savefig('country.png')
plt.savefig('country.pdf')



plt.show()
```
