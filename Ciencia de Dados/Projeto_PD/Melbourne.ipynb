Este [arquivo CSV](https://www.kaggle.com/datasets/anthonypino/melbourne-housing-market?select=MELBOURNE_HOUSE_PRICES_LESS.csv) contém dados do mercado imobiliário de Melbourne (Austrália).  
É um conjunto de dados sobre vendas de imóveis na cidade de Melbourne.



```python
import pandas as pd

import re
```

```python
#  DataFrame vai ser df
df = pd.read_csv('house.csv')
```

### Filtrando Dados com condições

```python
# Quartos: Room, Tipo = Type, h = Casa

filtro1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 500000)]

filtro1
```

### Filtrando Dados Regular Expression 

```python
#  Vamos fazer um filtro para as ruas, porém o emdereço está tudo jundo EX: 56A Turner St, logo ser for usado o == não vai dá certo.
# Para esse caso vou usar a função contei, ou seja no que eu quero filtrar tem que conter na linha e coluna esse pedasso que estou passando. 

filtro2_V1 = df.loc[df['Address'].str.contains('Turner')]

filtro2_V2 = df.loc[df['Address'].str.contains('Turner St')]

# filtro2 = df.loc[df['Address'].str.contains('Turner St | Turner Rd')]

# As busca de cima leveram em conta o kamiquase

filtro2 = df.loc[df['Address'].str.contains('Turner St | Turner Rd', flags=re.I)] # flags=re.I Vai despresar o kamiquase, o re tem que ser importado
```

### Filtrando Dados com Regx e Condição 

```python
# filtro3 = df.loc[df['Address'].str.contains('59', flags=re.I) & (df['Price'] <= 500000)] Aqui vão aprecer todas as cas com valor <+ 500000 e que tenham no numero o valor 59

filtro3 = df.loc[df['Address'].str.contains('^59', flags=re.I) & (df['Price'] <= 500000)] # Tem que iniciar com 59, graças ao sinal de ^ (tio)

# A casa tem que ser de n 59 e Price (Preço) Ihual ou menor que 500000

filtro3
```

### Modificando dados na DataFrame

```python
# Vamos filtar um vendedor
df.loc[df['SellerG'] == 'Nelson'] # Na coluna Vendedor (SellerG), Vamos achar o seu Nelson
```

```python
# Troca o nome do vendedor Nelson para Ronald
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronald' # Aqui busquei o Nelson na Coluna Seller e falei que SellerG recebe Ronald no lugar de Nelson
```

```python
# O Method de Ronal deve mudar de S para Pending
df.loc[df['SellerG'] == 'Ronald', ['Method']] = 'Pending' # OBS: Coluna sempre dentro de cocheites []
```

### Agrupando dados com o Groupby

```python
df.groupby(['SellerG']) # Aqui serve para agrupar os vendndores

# df
```

```python
# Vamos agrupar e ver algumasa funções.

df.groupby(['SellerG']).mean(numeric_only=True)

# A operação de média falha (TypeError) ao tentar ser aplicada em colunas de texto (dtype 'object').
# O parâmetro 'numeric_only=True' instrui o Pandas a ignorar essas colunas na agregação.
```
